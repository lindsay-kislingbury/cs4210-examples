```python
import pandas as pd
# dataframes
# load data from csv file
housing = pd.read_csv('./housing.csv')

# display the top rows
housing.head(10)
```

```python
# visualizing the data using a scatterplot
%matplotlib inline
import matplotlib.pyplot as plt

housing.plot(kind="scatter", x = "longitude", y = "latitude", alpha = 0.1)
plt.show()
```

```python
# get a quick description of the data, including # of rows, # of features, name of each feature, type of each feature, # of non-null values
housing.info()
```

```python
# get the series "ocean_proximity"
ocean_proximity_column = housing["ocean_proximity"]

# display the count of each unique category
ocean_proximity_column.value_counts()
```

```python
# show a summary of the numerical attributes
housing.describe()
```

```python
# show a histogram for each numerical attribute
# a histogram shows the number of instances that have a given value range
housing.hist(bins=50, figsize=(20,15))
plt.show()
```

# **Appendix I: List vs Pandas Frame vs Pandas Series vs Numpy Array**

*   https://www.geeksforgeeks.org/creating-a-dataframe-from-pandas-series/?ref=lbp

 Series can only contain a single list with an index, whereas Dataframe can be made of more than one series or we can say that a Dataframe is a collection of series that can be used to analyze the data

*   https://towardsdatascience.com/python-list-numpy-and-pandas-3a32f1aee948

1_rv1JADavAhDKN4-3iM7phQ.png

```python
import pandas as pd

# python lists
author = ['Jitender', 'Purnima', 'Arpit', 'Jyoti']
article = [210, 211, 114, 178]

# createe series
auth_series = pd.Series(author)
print('auth_series: ')
print(auth_series)
print()

article_series = pd.Series(article)
print('article_series: ')
print(article_series)
print()

# create a dataframe
frame = { 'Author': auth_series, 'Article': article_series}
result = pd.DataFrame(frame)
print('DataFrame:')
print(result)
print()

# add a column
age = [21, 21, 24, 23]
result['Age'] = pd.Series(age)
print(result)
print()

# add a row
new_row = {'Author': 'Machine', 'Article': 421, 'Age': 25}
# Use pd.concat instead of the deprecated append method
result = pd.concat([result, pd.DataFrame([new_row])], ignore_index=True)
print(result)
print()
```
